```python
#Check shape survived the slimming

m = tf.keras.models.load_model("disc.keras", compile=False)
print(m.input_shape, m.output_shape)

import numpy as np
out = m.predict(np.zeros((1, 256, 256, 5), dtype="float32"), verbose=0)
print(out.shape, float(out.min()), float(out.max()))

"""
What you're verifying:

shapes print as (None, 256, 256, 5) and (None, 256, 256, 1) — architecture survived the round-trip;
the prediction comes back (1, 256, 256, 1) — the forward pass runs;
min/max land inside [0, 1] — confirming the sigmoid output head is intact.
"""
```
